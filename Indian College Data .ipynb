```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
College=pd.read_csv('College_data.csv')
College
```

# Data Cleaning 

```python
College.isnull().sum()
```

```python
College.dtypes
```

```python
College.drop(College[College.Rating == '--'].index,inplace=True)
College.drop(College[College.Academic == '--'].index,inplace=True)
College.drop(College[College.Accommodation == '--'].index,inplace=True)
College.drop(College[College.Faculty == '--'].index,inplace=True)
College.drop(College[College.Infrastructure == '--'].index,inplace=True)
College.drop(College[College.Placement == '--'].index,inplace=True)
College.drop(College[College.Social_Life == '--'].index,inplace=True)
College.drop(College[College.College_Name == '--'].index,inplace=True)
College.drop(College[College.State == '--'].index,inplace=True)
College.drop(College[College.Stream == '--'].index,inplace=True)
College.drop(College[College.UG_fee == '--'].index,inplace=True)
College.drop(College[College.PG_fee == '--'].index,inplace=True)
College
```

```python
# Reset Index
College.reset_index(drop=True,inplace=True)
College
```

```python
College.dtypes
```

```python
# Converting the object datatype into float for numerical vaues
College["UG_fee"]=College["UG_fee"].str.replace(',','').astype('float64')
College["PG_fee"]=College["PG_fee"].str.replace(',','').astype('float64')
College["Placement"]=College["Placement"].astype("float64")
College["Rating"]=College["Rating"].astype("float64")
College["Accommodation"]=College["Accommodation"].astype("float64")
College["Academic"]=College["Academic"].astype("float64")
College["Faculty"]=College["Faculty"].astype("float64")
College["Social_Life"]=College["Social_Life"].astype("float64")
College["Infrastructure"]=College["Infrastructure"].astype("float64")
```
